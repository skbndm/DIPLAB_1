```python
#Size reduction
import matplotlib.pyplot as plt
import cv2
import numpy as np

 
try:
    img = cv2.imread("C:\\Users\\B-tech\\Desktop\\ALPSRP243410540-H2.2_UA\\MASK-ALPSRP243410540-H2.2_UA.tif") 
    (height, width) = img.shape[:2]    
    res = cv2.resize(img, (int(width / 2), int(height / 2)), interpolation=cv2.INTER_CUBIC)
    cv2.imwrite('result.tiff', res)
    plt.imshow(res)
   
 
except IOError:
    print('Error while reading files !!!')
```

```python
#Rotating the image
try:
    img = cv2.imread("C:\\Users\\B-tech\\Desktop\\ALPSRP243410540-H2.2_UA\\MASK-ALPSRP243410540-H2.2_UA.tif")
    (rows, cols) = img.shape[:2]
    M = cv2.getRotationMatrix2D((cols / 2, rows / 2), 55, 1)
    res1 = cv2.warpAffine(img, M, (cols, rows))
    cv2.imwrite('result.tiff', res1)
    plt.imshow(res1)
except IOError:
    print('Error while reading files !!!')
```

```python
# Shifting image shift by (600, 600).
M = np.float32([[1, 0, 200], [0, 1, 200]])
try:
    img = cv2.imread("C:\\Users\\B-tech\\Desktop\\ALPSRP243410540-H2.2_UA\\MASK-ALPSRP243410540-H2.2_UA.tif")
    res2 = cv2.warpAffine(img, M, (cols, rows))
    cv2.imwrite('result.tiff', res2)
    plt.imshow(res2)
except IOError:
    print('Error while reading files !!!')
```

```python
#Canny edge reduction
try:
    img = cv2.imread("C:\\Users\\B-tech\\Desktop\\ALPSRP243410540-H2.2_UA\\MASK-ALPSRP243410540-H2.2_UA.tif")
    edges = cv2.Canny(img, 5, 1)
    cv2.imwrite('result.tiff', edges)
    plt.imshow(edges)
except IOError:
    print('Error while reading files !!!')
```
